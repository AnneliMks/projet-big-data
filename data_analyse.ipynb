```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
df_final = pd.read_csv("bases_fusionnees.csv")
```

```python
df_final.columns
```

```python
#Répartition du type de biens 
fig_type = px.pie(df_final, names='type_bien', 
                 title="Répartition du type de biens en vente dans le Rhône",
                 hole=0.4,
                 color_discrete_sequence=px.colors.qualitative.Pastel)
fig_type.show()
```

```python
#distribution des prix
fig_dist_prix = px.histogram(df_final, x="prix", 
                            nbins=50, 
                            title="Distribution des prix de vente",
                            labels={'prix': 'Prix (€)'},
                            color_discrete_sequence=['#636EFA'])
fig_dist_prix.show()
```

```python
median = np.median(df_final["prix"])
moyenne = np.mean(df_final["prix"])
print(f"Médiane des prix : {median.round()}")
print(f"Moyenne des prix {moyenne.round(2)}")
```

```python
all_villes = df_final.groupby('ville')['prix'].mean().reset_index()

#top10 des villes
top_10_villes = all_villes.sort_values(ascending=False, by='prix').head(10)

#+Lyon
lyon_data = all_villes[all_villes['ville'].str.contains('Lyon', case=False, na=False)].head(1)

final_plot_df = pd.concat([top_10_villes, lyon_data]).drop_duplicates().sort_values(by='prix', ascending=False)

fig_villes = px.bar(final_plot_df, x='ville', y='prix',
                   title="Top 10 des villes les plus chères + Lyon (Prix Moyen)",
                   text_auto='.2s',
                   color='prix',
                   color_continuous_scale='Viridis',
                   labels={'prix': 'Prix Moyen (€)', 'ville': 'Ville'})


fig_villes.show()
```

```python
#top 10 des maisons et des appartement les plus chers
top_maisons = df_final[df_final['type_bien'].str.contains('maison', case=False)].nlargest(10, 'prix')
top_apparts = df_final[df_final['type_bien'].str.contains('appart', case=False)].nlargest(10, 'prix')
top_biens = pd.concat([top_maisons, top_apparts])

fig_top = px.scatter(top_biens, x="ville", y="prix", color="type_bien", 
                    size="surface",
                    title="Top 10 des maisons/appartements les plus chers",
                    hover_data=['surface', 'pieces'])
fig_top.show()
```

```python
#prix au m² par ville
df_final['prix_m2'] = df_final['prix'] / df_final['surface']
top_m2 = df_final.groupby('ville')['prix_m2'].mean().sort_values(ascending=False).head(10).reset_index()

fig_m2 = px.bar(top_m2, x='ville', y='prix_m2',
               title="Top 10 des villes avec le prix au m² le plus élevé",
               labels={'prix_m2': 'Prix moyen au m² (€)'},
               color_discrete_sequence=['#EF553B'])
fig_m2.show()
```

```python
#Prix en fonction de la surface
fig = px.scatter(df_final, x="surface", y="prix", 
                 color="type_bien",
                 title="Prix vs Surface dans le Rhône",
                 hover_name="ville")

fig.show()
```
